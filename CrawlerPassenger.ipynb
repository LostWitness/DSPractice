#### 1.1 Importing BeautifulSoup library to parse the content from a given website

```python
from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen, Request

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
reg_url = "https://passengermusic.com/live/"
req = Request(url=reg_url, headers=headers) 
html = urlopen(req).read() 
#print(html) 
#url = "https://passengermusic.com/live/"
#html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Step 2: Parse the html content
import pprint
# Print out the text
text = soup.get_text()
soup = BeautifulSoup(text, "lxml")
# print the parsed data of html
print(soup.prettify())
```

```python
rows_dates = soup.find_all(class_='date')
rows_venues = soup.find_all(class_='venue')
rows_locations = soup.find_all(class_='location')
#print(rows_dates)
```
